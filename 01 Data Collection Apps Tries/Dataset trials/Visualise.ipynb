## Preprocess motion sensor data.
### It includes:
- Normalization: Using Min-Max scaling for accelerometer and gyroscope values.
- Segmentation: Sliding windows of 3 seconds (assuming timestamps are in milliseconds).
- Feature Extraction: Mean, variance, and FFT magnitude for each segment.

```python
import pandas as pd
import numpy as np
from scipy.fft import fft
from sklearn.preprocessing import MinMaxScaler

# Load data
df = pd.read_csv('hs_EE-C1.csv')  # Change to your filename

# Normalize accelerometer and gyroscope values
scaler = MinMaxScaler()
df[['X_Acc', 'Y_Acc', 'Z_Acc', 'X_Gyro', 'Y_Gyro', 'Z_Gyro']] = scaler.fit_transform(
    df[['X_Acc', 'Y_Acc', 'Z_Acc', 'X_Gyro', 'Y_Gyro', 'Z_Gyro']]
)

# Convert Timestamp to seconds
df['Timestamp'] = df['Timestamp'] / 1000  # Convert ms to s

# Segmenting the data into 3-second windows
window_size = 3  # seconds
segments = []
labels = []

start_idx = 0
while start_idx < len(df):
    end_time = df.loc[start_idx, 'Timestamp'] + window_size
    segment = df[(df['Timestamp'] >= df.loc[start_idx, 'Timestamp']) & 
                 (df['Timestamp'] < end_time)]
    
    if len(segment) > 0:
        # Compute features for the segment
        features = {
            'mean_x_acc': segment['X_Acc'].mean(),
            'var_x_acc': segment['X_Acc'].var(),
            'mean_y_acc': segment['Y_Acc'].mean(),
            'var_y_acc': segment['Y_Acc'].var(),
            'mean_z_acc': segment['Z_Acc'].mean(),
            'var_z_acc': segment['Z_Acc'].var(),
            'mean_x_gyro': segment['X_Gyro'].mean(),
            'var_x_gyro': segment['X_Gyro'].var(),
            'mean_y_gyro': segment['Y_Gyro'].mean(),
            'var_y_gyro': segment['Y_Gyro'].var(),
            'mean_z_gyro': segment['Z_Gyro'].mean(),
            'var_z_gyro': segment['Z_Gyro'].var(),
        }
        
        # FFT features
        for axis in ['X_Acc', 'Y_Acc', 'Z_Acc', 'X_Gyro', 'Y_Gyro', 'Z_Gyro']:
            fft_values = np.abs(fft(segment[axis].values))[:len(segment)//2]  # Magnitude spectrum
            features[f'fft_{axis}_max'] = np.max(fft_values)
            features[f'fft_{axis}_mean'] = np.mean(fft_values)
        
        segments.append(features)
    
    start_idx += len(segment)  # Move to the next window

# Convert to DataFrame
processed_data = pd.DataFrame(segments)

# Save processed data
processed_data.to_csv("processed_sensor_data.csv", index=False)

print("Data preprocessing complete. Processed file saved as 'processed_sensor_data.csv'.")
```

# Data Preprocessing for Sensors
### **Yaw, Pitch, and Roll Definitions**

1.  **Pitch (θ)**: Rotation around the **X-axis** (tilt forward or backward).
2.  **Roll (ϕ)**: Rotation around the **Y-axis** (tilt sideways).
3.  **Yaw (ψ)**: Rotation around the **Z-axis** (compass heading).

We compute them using **accelerometer and gyroscope data** as follows:

-   **From Accelerometer (Static Orientation)**:
    
    **Pitch Equation:**

    $$ \text{Pitch} = \arctan \left( \frac{Y_{\text{Acc}}}{\sqrt{X_{\text{Acc}}^2 + Z_{\text{Acc}}^2}} \right) $$

    **Roll Equation:**

    $$ \text{Roll} = \arctan \left( \frac{-X_{\text{Acc}}}{Z_{\text{Acc}}} \right) $$

    -   No direct yaw calculation from the accelerometer.
-   **From Gyroscope (Dynamic Change)**:
    
    -   Integrate gyroscope readings to estimate YPR changes over time.


### **📌 How This Works?**

-   **Pitch & Roll** come from the **accelerometer** (static orientation).
-   **Yaw** is **integrated** from **Z-axis gyroscope** values.
-   **Visualization** helps track how the phone moves over time.

This method works well for short durations. For longer tracking, you'll need **sensor fusion** (e.g., **Kalman Filter or Complementary Filter**) to reduce drift.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv('hs_EE-C1.csv')  # Change filename if needed

# Convert Timestamp to seconds
df['Timestamp'] = df['Timestamp'] / 1000  # Convert ms to seconds

# Convert accelerometer readings to Pitch and Roll (in degrees)
df['Pitch'] = np.degrees(np.arctan2(df['Y_Acc'], np.sqrt(df['X_Acc']**2 + df['Z_Acc']**2)))
df['Roll'] = np.degrees(np.arctan2(-df['X_Acc'], df['Z_Acc']))

# Yaw Calculation (Using Gyroscope - Integration)
df['Delta_Time'] = df['Timestamp'].diff().fillna(0)  # Time difference in seconds
df['Yaw'] = np.degrees(np.cumsum(df['Z_Gyro'] * df['Delta_Time']))  # Integrate Z_Gyro over time

# Plot Pitch, Roll, Yaw over time
plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['Pitch'], label='Pitch', color='r')
plt.plot(df['Timestamp'], df['Roll'], label='Roll', color='g')
plt.plot(df['Timestamp'], df['Yaw'], label='Yaw', color='b')
plt.xlabel("Time (s)")
plt.ylabel("Angle (degrees)")
plt.title("Yaw, Pitch, and Roll Over Time")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv('hs_EE-C1.csv')  # Change filename if needed

# Convert Timestamp to seconds
df['Timestamp'] = df['Timestamp'] / 1000  # Convert ms to sec
df['Delta_Time'] = df['Timestamp'].diff().fillna(0)  # Compute time difference

# Accelerometer-based Pitch & Roll
df['Pitch_Accel'] = np.degrees(np.arctan2(df['Y_Acc'], np.sqrt(df['X_Acc']**2 + df['Z_Acc']**2)))
df['Roll_Accel'] = np.degrees(np.arctan2(-df['X_Acc'], df['Z_Acc']))

# Initialize Filtered Angles
df['Pitch'], df['Roll'], df['Yaw'] = 0, 0, 0
alpha = 0.98  # Complementary filter parameter

# Complementary Filter Loop
for i in range(1, len(df)):
    dt = df.loc[i, 'Delta_Time']
    
    # Gyroscope-based estimation
    gyro_pitch = df.loc[i-1, 'Pitch'] + df.loc[i, 'X_Gyro'] * dt
    gyro_roll = df.loc[i-1, 'Roll'] + df.loc[i, 'Y_Gyro'] * dt
    gyro_yaw = df.loc[i-1, 'Yaw'] + df.loc[i, 'Z_Gyro'] * dt  # No accel correction for yaw

    # Complementary filter
    df.loc[i, 'Pitch'] = alpha * gyro_pitch + (1 - alpha) * df.loc[i, 'Pitch_Accel']
    df.loc[i, 'Roll'] = alpha * gyro_roll + (1 - alpha) * df.loc[i, 'Roll_Accel']
    df.loc[i, 'Yaw'] = gyro_yaw  # Gyro-only for yaw

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['Pitch'], label='Pitch', color='r')
plt.plot(df['Timestamp'], df['Roll'], label='Roll', color='g')
plt.plot(df['Timestamp'], df['Yaw'], label='Yaw', color='b')
plt.xlabel("Time (s)")
plt.ylabel("Angle (degrees)")
plt.title("Yaw, Pitch, and Roll (Complementary Filter)")
plt.legend()
plt.show()
```

## Yaw Pitch Roll Kalman Filter

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv('hs_EE-C1.csv')

# Convert Timestamp to seconds
df['Timestamp'] = df['Timestamp'] / 1000
df['Delta_Time'] = df['Timestamp'].diff().fillna(0)

# Accelerometer-based Pitch & Roll
df['Pitch_Accel'] = np.degrees(np.arctan2(df['Y_Acc'], np.sqrt(df['X_Acc']**2 + df['Z_Acc']**2)))
df['Roll_Accel'] = np.degrees(np.arctan2(-df['X_Acc'], df['Z_Acc']))

# Initialize state (pitch, roll, yaw) and covariance matrix
state = np.array([[0], [0], [0]])  # Initial angles: Pitch, Roll, Yaw
P = np.eye(3) * 0.1  # Initial uncertainty

# Process and measurement noise
Q = np.eye(3) * 0.01  # Process noise (gyroscope noise)
R = np.eye(3) * 0.1   # Measurement noise (accelerometer noise)

# Identity matrix
I = np.eye(3)

# State transition matrix (based on gyroscope integration)
A = np.eye(3)

# Measurement matrix (relates accelerometer readings to pitch & roll)
H = np.eye(3)

# Kalman Filter loop
filtered_states = []

for i in range(1, len(df)):
    dt = df.loc[i, 'Delta_Time']
    
    # Gyroscope-based prediction (Euler integration)
    gyro_input = np.array([[df.loc[i, 'X_Gyro'] * dt], 
                           [df.loc[i, 'Y_Gyro'] * dt], 
                           [df.loc[i, 'Z_Gyro'] * dt]])
    
    # Predict state
    state = A @ state + gyro_input
    P = A @ P @ A.T + Q  # Update uncertainty

    # Measurement (Accelerometer)
    measurement = np.array([[df.loc[i, 'Pitch_Accel']], 
                            [df.loc[i, 'Roll_Accel']], 
                            [state[2, 0]]])  # No direct yaw measurement

    # Kalman Gain
    S = H @ P @ H.T + R
    K = P @ H.T @ np.linalg.inv(S)

    # Update step
    state = state + K @ (measurement - H @ state)
    P = (I - K @ H) @ P  # Update uncertainty

    # Store results
    filtered_states.append(state.flatten())

# Convert list to DataFrame
filtered_states = np.array(filtered_states)
df = df.iloc[1:]  # Adjust df to match the length of filtered_states
df['Pitch_KF'] = filtered_states[:, 0]
df['Roll_KF'] = filtered_states[:, 1]
df['Yaw_KF'] = filtered_states[:, 2]

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['Pitch_KF'], label='Pitch (Kalman)', color='r')
plt.plot(df['Timestamp'], df['Roll_KF'], label='Roll (Kalman)', color='g')
plt.plot(df['Timestamp'], df['Yaw_KF'], label='Yaw (Kalman)', color='b')
plt.xlabel("Time (s)")
plt.ylabel("Angle (degrees)")
plt.title("Yaw, Pitch, and Roll (Kalman Filter)")
plt.legend()
plt.show()
```
